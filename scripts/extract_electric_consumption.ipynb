```python
import pandas as pd
from pathlib import Path

DATA_URL = "https://data.cityofnewyork.us/api/views/jr24-e7cr/rows.csv?accessType=DOWNLOAD"

OUTPUT_DIR = Path("../data_landing")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

OUTPUT_FILE = OUTPUT_DIR / "electric_consumption_2010_2025_raw.csv"

df = pd.read_csv(DATA_URL)

df.head()
```

```python
df.to_csv(OUTPUT_FILE, index=False)
OUTPUT_FILE
```
